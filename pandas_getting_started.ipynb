### What is a Series?

##### A Pandas Series is like a column in a table. It is a one-dimensional array holding data of any type.


```python
import pandas as pd

x = [1,7,2]

myvar = pd.Series(x)

print(myvar)
```

### Create Labels
With the index argument, you can name your own labels.

```python
# we can replace above 0 1 2 serials with own numbering example :

import pandas as pd

x =[5,6,7]

myvar = pd.Series(x,index = ["x","y","z"])

print(myvar)
```

```python
print(myvar["y"])
```

## Key/Value Objects as Series

#### Note: The keys of the dictionary become the labels.

```python
calories = {"day1": 420, "day2": 380, "day3": 390}
```

```python
import pandas as pd

myvar = pd.Series(calories)

print(myvar)
```

### if we add custom label of only day1 and day2 only data of day1 & day 2 will come and other remainig will vanish :

```python
import pandas as pd

myvar = pd.Series(calories,index = ["day1","day2"])

print(myvar)
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
```

```python
import pandas as pd

myvar = pd.DataFrame(data)

print(myvar)
```

## Locate Row
As you can see from the result above, the DataFrame is like a table with rows and columns.
Pandas use the loc attribute to return one or more specified row(s)

```python
#refer to the row index:
print(myvar.loc[0])
```

### here we located the row ZERO and python returned its values

```python
#use a list of indexes:
print(myvar.loc[[0, 1]])
```

```python
# NOW lets add a custom index to data

myvar = pd.DataFrame(data,index=['day1','day2','day3'])
```

# Pandas - Analyzing DataFrames

```python
import pandas as pd

df = pd.read_csv('data.csv')

print(df.head(10))
```
